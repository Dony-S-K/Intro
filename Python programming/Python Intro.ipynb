## Example 1: Python Program to add two numbers

```python
a = 4
b = 2
sum = a + b
print(sum)
```

```python
Num1 = float(input("Enter the First Number: "))
Num2 = float(input("Enter the Second Number: "))

def sum(num1, num2):
    sum = num1 + num2
    return sum

print("Sum of two numbers: ", sum(Num1, Num2))
```

## Example 2: Python Program that Prints First N natural numbers

```python
Num = int(input("Enter the Number: "))
for i in range(1, Num+1):
    print(i)
    
```

## Example 3: Python Program to check whether a number is Even or Odd

```python
Num = int(input("Enter the Number: "))

if Num % 2 == 0:
    print(Num, " is an Even Number")
else:
    print(Num, " is an Odd Number")
```

## Example 4: Python Program to find Sum and Average of First N natural numbers

```python
Num = int(input("Enter the Natural Number: "))
sum = 0
for i in range(1, Num + 1):
    sum = sum + i
avr = sum/Num  
print("Sum of First", Num,"Natural number is:", sum)
print("Average of First", Num,"Natural number is: ", avr)
```

## Example 5: Program that Finds the Largest Among Three Numbers

```python
Num1 = float(input("Enter the First Number: "))
Num2 = float(input("Enter the Second Number: "))
Num3 = float(input("Enter the Third Number: "))

if Num1>Num2>Num3:
    print("The largest among three numbers is: ", Num1)
elif Num2>Num3:
    print("The largest among three numbers is: ", Num2)
else:
    print("The largest among three numbers is: ", Num3)
     
```
